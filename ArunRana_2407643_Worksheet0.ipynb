```python
#Task1:
def convert_units(conversion_type, value):
    """
    Convert a value from one unit to another.
    Parameters
    conversion_type: type of conversion
    value: number to convert
    Returns:
    Converted value.
    """

    if conversion_type == "m_to_ft":
        return value * 3.28084
    elif conversion_type == "ft_to_m":
        return value / 3.28084
    elif conversion_type == "kg_to_lbs":
        return value * 2.20462
    elif conversion_type == "lbs_to_kg":
        return value / 2.20462
    elif conversion_type == "l_to_gal":
        return value * 0.264172
    elif conversion_type == "gal_to_l":
        return value / 0.264172
    else:
        raise ValueError("Unsupported conversion type.")


def main():
    print("Unit Conversion Program")
    print("1. Length")
    print("2. Weight")
    print("3. Volume")

    try:
        category = int(input("Enter choice (1-3): "))

        if category == 1:
            print("a. Meters to Feet")
            print("b. Feet to Meters")
            option = input("Choose (a/b): ").lower()
            conversion_type = "m_to_ft" if option == "a" else "ft_to_m"

        elif category == 2:
            print("a. Kilograms to Pounds")
            print("b. Pounds to Kilograms")
            option = input("Choose (a/b): ").lower()
            conversion_type = "kg_to_lbs" if option == "a" else "lbs_to_kg"

        elif category == 3:
            print("a. Liters to Gallons")
            print("b. Gallons to Liters")
            option = input("Choose (a/b): ").lower()
            conversion_type = "l_to_gal" if option == "a" else "gal_to_l"

        else:
            raise ValueError("Invalid category.")

        value = float(input("Enter value: "))
        result = convert_units(conversion_type, value)

        print("Converted value:", round(result, 4))

    except ValueError:
        print("Invalid input! Please enter correct values.")


if __name__ == "__main__":
    main()
```

```python
#Task2:
def calculate_sum(numbers):
    """Return the sum of numbers in the list."""
    return sum(numbers)


def calculate_average(numbers):
    """Return the average of numbers in the list."""
    return sum(numbers) / len(numbers)


def find_maximum(numbers):
    """Return the largest number in the list."""
    return max(numbers)


def find_minimum(numbers):
    """Return the smallest number in the list."""
    return min(numbers)


def main():
    print("Mathematical Operations")
    print("1. Sum")
    print("2. Average")
    print("3. Maximum")
    print("4. Minimum")

    try:
        choice = int(input("Choose an operation (1-4): "))
        user_input = input("Enter numbers separated by spaces: ")

        number_list = [float(num) for num in user_input.split()]

        if len(number_list) == 0:
            raise ValueError("Empty list")

        if choice == 1:
            print("Sum:", calculate_sum(number_list))
        elif choice == 2:
            print("Average:", calculate_average(number_list))
        elif choice == 3:
            print("Maximum:", find_maximum(number_list))
        elif choice == 4:
            print("Minimum:", find_minimum(number_list))
        else:
            print("Invalid choice.")

    except ValueError:
        print("Invalid input! Please enter numbers only.")


if __name__ == "__main__":
    main()
```

```python
#1
def extract_every_other(lst):
    """Return every other element from the list starting from the first element."""
    return lst[::2]
numbers = [1, 2, 3, 4, 5, 6]
result = extract_every_other(numbers)
print(result)
```

```python
#2
def get_sublist(lst, start, end):
    """Return a sublist from start index to end index (inclusive)."""
    return lst[start:end + 1]

numbers = [1, 2, 3, 4, 5, 6]
result = get_sublist(numbers, 2, 4)
print(result)
```

```python
#3
def reverse_list(lst):
    """Return the reversed list using slicing."""
    return lst[::-1]
numbers = [1, 2, 3, 4, 5]
result = reverse_list(numbers)
print(result)
```

```python
#4
def remove_first_last(lst):
    """Return a sublist without the first and last elements."""
    return lst[1:-1]
numbers = [1, 2, 3, 4, 5]
result = remove_first_last(numbers)
print(result)
```

```python
#5
def get_first_n(lst, n):
    """Return the first n elements from the list."""
    return lst[:n]
numbers = [1, 2, 3, 4, 5]
result = get_first_n(numbers, 3)
print(result)
```

```python
#Problem:4 Matrix Operation:
import numpy as np

A = np.array([[3, 4], [7, 8]])
B = np.array([[5, 3], [2, 1]])

# 1. Prove A * A⁻¹ = I
A_inv = np.linalg.inv(A)
identity_check = np.dot(A, A_inv)
print("A * A⁻¹ = Identity Matrix:\n", identity_check)

# 2. Prove AB ≠ BA
AB = np.dot(A, B)
BA = np.dot(B, A)
print("\nMatrix AB:\n", AB)
print("\nMatrix BA:\n", BA)
if not np.array_equal(AB, BA):
    print("\nAB is NOT equal to BA")

# 3. Prove (AB)ᵀ = Bᵀ * Aᵀ
AB_transpose = np.transpose(AB)
BT_AT = np.dot(B.T, A.T)
print("\n(AB)ᵀ:\n", AB_transpose)
print("\nBᵀ * Aᵀ:\n", BT_AT)
if np.array_equal(AB_transpose, BT_AT):
    print("\n(AB)ᵀ is equal to Bᵀ * Aᵀ")

# 4. Solve Linear Equations using Inverse Methods
# System of equations:
#  2x - 3y + z = -1
#  x - y + 2z = -3
#  3x + y - z = 9

A_matrix = np.array([[2, -3, 1],
                     [1, -1, 2],
                     [3, 1, -1]])

B_matrix = np.array([-1, -3, 9])

A_inv = np.linalg.inv(A_matrix)
solution = np.dot(A_inv, B_matrix)
print("\nSolution for Linear Equations (x, y, z):\n", solution)
```
